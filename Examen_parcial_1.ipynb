```python
#8
data = pd.DataFrame(np.arange(12).reshape((3, 4)),index=['CDMX', 'Guanajuato', 'Puebla'],columns=['uno', 'dos', 'tres', 'cuatro'])
data
```

La parte del index y la parte de columns son para indicar qúe títulos queremos que lleven las filas y las columnas del data frame
