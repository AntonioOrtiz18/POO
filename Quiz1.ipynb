##Diccionario con 3 llaves

```python
d= {"Mate":[8,10,9,7.6,5.4],
              "Español":[5,8,7,9.3,4.2],
              "Inglés":[3,7,6,9.4,8.8]
}
#creación del diccionario
```

```python
d
```

##Mostrar sus items y el tipo de objeto que se obtiene


```python
i=d.items()
i
#Muestra los items del diccionario creado
```

```python
type(i)
#muestra el tipo de objeto que es
```

##Panel de datos usando el diccionario

```python
import pandas as pd
#importa el módulo para crear el data frame
```

```python
df=pd.DataFrame(d)
#Hace el diccionario un data frame
```

```python
df
```

##3 segregaciones del diccionario

```python
d1=df.iloc[2:4]
#Muestra la fila 3 y 4 del data frame
```

```python
d1
```

```python
d2=df.drop(columns="Inglés")
#Esta segregación quita del data frame la columna con el nombre Inglés
```

```python
d2
```

```python
d3=df.iloc[:,[0,2]]
#muestra solo las columnas 1 y 3(o 0 y 2) del data frame
```

```python
d3
```

##Análisis descriptivo panel de datos

```python
df.describe()
#Saca la estadística del data frame
```

Podemos observar que el promedio de matemáticas fue el más alto, con 8. El promedio más bajo fue en español con 6.7. La calificación más baja en matemáticas fue de 5.4, mientras que la más alta fue de 10; en español la más baja fue de 4.2, mientras que la más alta fue de 9.3; en inglés la más baja fue de 3, mientras que la más alta fue 9.4. La mediana en español e inglés fue igual con 7.
